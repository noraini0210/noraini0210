### **Working on multiple files**

```python
#Specify to display plots under the code cell block
%matplotlib inline
```

```python
#Combine multiple OHC data from the drive
ohcw = xr.open_mfdataset('*.nc',parallel=True)
```

There were **972 .nc files (monthly OHC data from 1940 to 2020)** in the folder 

Systematic methods required to read all the files is 
shown above. The function automatically concatenates and merges multiple files into a single xarray dataset. 

Time is represented as "yyyymm"

```python

ohcw
```

```python
ohcw.OHC100
```

```python
ohcw.OHC100.isel(time=12).plot(x="lon")
```

`isel `is in contrast to `.sel` which allows you to select data by value.

```python
ohcw0 = ohcw.isel(time=0)
ohcw0.lat.plot()
ohcw0.lat.plot()
```

```python
#Visualizing all OHC in January 1940 at different depths
fig, axes = plt.subplots(2,3,figsize=(20,10))
ohcw0.OHC100.plot(ax=axes[0,0])
axes[0,0].set(title='1940001 OHC (0-100m)')
ohcw0.OHC300.plot(ax=axes[0,1])
axes[0,1].set(title='1940001 OHC (0-300m)')
ohcw0.OHC700.plot(ax=axes[0,2])
axes[0,2].set(title='1940001 OHC (0-700m)')
ohcw0.OHC1500.plot(ax=axes[1,0])
axes[1,0].set(title='1940001 OHC (0-1500m)')
ohcw0.OHC2000.plot(ax=axes[1,1])
axes[1,1].set(title='1940001 OHC (0-2000m)')
fig.tight_layout()
plt.show()
```

### **Visualize OHC over IPWP**

```python
#Try 1-month data
#IPWP area: (60E - 100W and 30S - 30N)
ohcw0_IPWP = ohcw0.where((ohcw0.lon>59.5) & (ohcw0.lon<280.5) & (ohcw0.lat>-30.5) & (ohcw0.lat<30.5),drop=True)
ohcw0_IPWP 
```

```python
#Visualizing all OHC in January 1940 at different depths
fig, axes = plt.subplots(2,3,figsize=(20,10))
ohcw0_IPWP.OHC100.plot(ax=axes[0,0])
axes[0,0].set(title='1940001 OHC (0-100m)')
ohcw0_IPWP.OHC300.plot(ax=axes[0,1])
axes[0,1].set(title='1940001 OHC (0-300m)')
ohcw0_IPWP.OHC700.plot(ax=axes[0,2])
axes[0,2].set(title='1940001 OHC (0-700m)')
ohcw0_IPWP.OHC1500.plot(ax=axes[1,0])
axes[1,0].set(title='1940001 OHC (0-1500m)')
ohcw0_IPWP.OHC2000.plot(ax=axes[1,1])
axes[1,1].set(title='1940001 OHC (0-2000m)')
fig.tight_layout()
plt.show()
```
